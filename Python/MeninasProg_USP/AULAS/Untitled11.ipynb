```python
cpf = input()

cpf_novo = ''
alg = '0123456789'
enconding = 0


for i in cpf:
    if i != '.' and i != '-':
         cpf_novo += i
         if i not in alg:
           enconding = 1


if enconding == 1:
  print('ENCONDING ERROR')
if len(cpf_novo) != 11:
  print('SIZE ERROR')
if enconding == 0 and len(cpf_novo) == 11:
  print(cpf_novo)


```
