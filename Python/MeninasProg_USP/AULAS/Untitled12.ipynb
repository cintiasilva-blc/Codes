```python
num_votos = int(input())

favor = 0
contra = 0
outros = 0

for i in range(num_votos):
  voto = input()
  if voto == 'F':
    favor += 1
  if voto == 'C':
    contra += 1
  if voto == 'A' or voto == 'B' or voto == 'N':
    outros += 1

print(f'a favor: {favor}')
print(f'conta: {contra}')
print(f'outros: {outros}')
```
