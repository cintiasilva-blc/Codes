```python
lin1 = input()
lin2 = input()
result = ''

for it in lin1:
  if it in lin2:
    result += '*'
  else:
    result += it

print(result)
```

```python
min = int(input())
vlrs = 0

while min > vlrs:
  vlr = int(input())
  vlrs += vlr

sobra = vlrs - min

print(f'minimo {min}')
print(f'total {vlrs}')
print(f'sobra {sobra}')
```

```python
parado = 1
movimento = 0
dado = input()

while dado != 'f':
   if dado == 'p':
      parado += 1
   elif dado == 'm':
      movimento += 1
   dado = input()

if parado < movimento:
  print('ativo')
else:
  print('sedentário')
```

```python
palavra = input()
let = ''

while True:
    carc = input()
    if carc == '.':
       break
    let += carc

if let == palavra:
  print('True')
else:
  print('False')
```

```python
cal_recom = int(input())
cosm = 0

while True:
  if cosm < cal_recom:
    vl_cosm = int(input())
    cosm += vl_cosm
  else:
    break

print(f'total: {cosm}')
```
